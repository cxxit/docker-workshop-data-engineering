```python
dtype = {
    "LocationID": "Int64",
    "Borough": "string",
    "Zone" : "string",
    "service_zone" : "string"
}


df = pd.read_csv(
    url,
    dtype=dtype
)
```

```python
df.dtypes
```

```python
# Create a connection to the PostgreSQL database using SQLAlchemy
from sqlalchemy import create_engine
engine = create_engine('postgresql+psycopg://root:root@localhost:5432/ny_taxi')
```

```python
print(pd.io.sql.get_schema(df, name='taxi_zone_lookup', con=engine))
```

```python
# just creating the schema an empty dataframe
df.head(n=0).to_sql(name='taxi_zone_lookup', con=engine, if_exists='replace')
```

```python
df_iter = pd.read_csv(
    url,
    dtype=dtype,
    iterator=True,
    chunksize=10)
```

```python
df = next(df_iter)
```

```python
df
```

```python
from tqdm.auto import tqdm
```

```python
for df_chunk in tqdm(df_iter):
    df_chunk.to_sql(name='taxi_zone_lookup', con=engine, if_exists='append')
```
